Write a Pandas program to filter the records from the world alcohol consumption dataset where the GDP per capita is above 10,000 and the wine servings are greater than 100. Retrieve the country, beer servings, spirit servings, and life expectancy for these filtered records.

```python
import pandas as pd

df = pd.read_csv("dataset/World_Alcohol_Dataset - MAIN.csv")
```

```python
df
```

```python
df[df["wine_servings"].str.contains("\?")] #we can see the row with the "?" character.
```

```python
df=df[df["wine_servings"]!="?"] #we dropped the row due the missing data on wine_servings
```

```python
df[df["wine_servings"].str.contains("\?")] #we check again
```

```python
df["wine_servings"]=df["wine_servings"].astype(int) #we cast the int type on "wine_servings"
```

```python
df.dtypes
```

```python
filtered_df=df[(df["gdp_per_capita"]>10000)&(df["wine_servings"]>100)] #we initialize another df to filter the requested infos

filtered_df[["country","beer_servings","spirit_servings","life_expentancy"]] #we visualize only the requested columns
```
