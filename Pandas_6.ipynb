Write a Pandas program to find the top 5 countries with the highest beer servings and the highest spirit servings from the world alcohol consumption dataset. Retrieve the country names, beer servings, spirit servings, and the corresponding region for these top 5 countries.

```python
import pandas as pd

df=pd.read_csv("dataset/World_Alcohol_Dataset - MAIN.csv")
```

```python
df1=df.copy() #defining a new dataframe 
df1=df1[df1["wine_servings"]!="?"]
df1
```

```python
df1=df1[df1["beer_servings"]!="?"]
df1=df1[df1["spirit_servings"]!="?"]
```

```python
df1["beer_servings"]=df1["beer_servings"].astype(int)
df1["wine_servings"]=df1["wine_servings"].astype(int)
df1["spirit_servings"]=df1["spirit_servings"].astype(int)
```

```python
df1.info()
```

```python
top5beer= df1.sort_values(by="beer_servings", ascending=False).head(5)
top5beer=top5beer[["country","region","beer_servings"]]
top5beer
```

```python
top5spirit=df1.sort_values(by="spirit_servings", ascending=False).head(5)
top5spirit= top5spirit[["country","region","spirit_servings"]]
top5spirit
```
