Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframeâ€™s specific columns with different colors.

```python
import pandas as pd
import numpy as np

# Creating the df
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['first', 'second', 'third', 'fourth']) # Converting the array in a df
```

```python
df
```

```python
def highlight_columns(dfcolumns, columns_to_highlight) :
    
    columns_to_highlight = columns_to_highlight.split(',') # prepare the user input

    for column in columns_to_highlight :
        if column.lower() in dfcolumns.name.lower():
            return ['background-color: green'] * len(dfcolumns)
        
    return [''] * len(dfcolumns)

# User input to choose the columns to highlight
input = input('Choose columns to highlight separated by a comma: ')

# Using 'highlight_columns' function and user input
coloured_df = df.style.apply(highlight_columns, columns_to_highlight=input)
coloured_df
```
