```python
import pandas as pd
```

```python
student_data = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
    index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
student_data.head(10)
```

```python
MEAN_numeric_df=student_data.groupby(by="school_code")["age"].mean().reset_index()
MEAN_numeric_df=MEAN_numeric_df.rename(columns={"age":"avg_age"})
MEAN_numeric_df
```

```python
MAX_numeric_df=student_data.groupby(by="school_code")["age"].max().reset_index()
MAX_numeric_df=MAX_numeric_df.rename(columns={"age":"max_age"})
MAX_numeric_df
```

```python
MIN_numeric_df=student_data.groupby(by="school_code")["age"].min().reset_index()
MIN_numeric_df=MIN_numeric_df.rename(columns={"age":"min_age"})
MIN_numeric_df
```
