Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
import pandas as pd
import numpy as np
```

```python
# Creating the df
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['first', 'second', 'third', 'fourth']) # Converting the array in a df

#Setting values to nan
df.iloc[0, 1] = np.nan
df.iloc[1, 2] = np.nan
df.iloc[9, 3] = np.nan
df.iloc[6, 2] = np.nan
df.iloc[3, 1] = np.nan
df.iloc[9, 3] = np.nan

# Showing the df
df
```

```python
def spotnan(x):

    for x in df:
        if pd.isna(x):
            return ["background-color: green"]
        else:
            return ""

highlighted_df= df.style.applymap(spotnan)
```
