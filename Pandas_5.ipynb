Write a Pandas program to find the countries in the world alcohol consumption dataset where the beer servings, spirit servings, and wine servings are all greater than 50 liters. Retrieve the country names for these records.

```python
import pandas as pd
```

```python
df=pd.read_csv("dataset/World_Alcohol_Dataset - MAIN.csv")
```

```python
df.head()
```

```python
df1=df.copy() #defining a new dataframe 
df1=df1[df1["wine_servings"]!="?"]
df1
```

```python
df1=df1[df1["beer_servings"]!="?"]
df1=df1[df1["spirit_servings"]!="?"]
```

```python
#changing types of the three columns
df1["beer_servings"]=df1["beer_servings"].astype(int)
df1["wine_servings"]=df1["wine_servings"].astype(int)
df1["spirit_servings"]=df1["spirit_servings"].astype(int)
```

```python
df1.info()
```

```python
df1[(df1["beer_servings"]>50)&(df1["wine_servings"]>50)&(df1["spirit_servings"]>50)]["country"].to_frame()
```
